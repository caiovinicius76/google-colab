## Implementar os itens abaixo:

#### b-) Ler o dataset "*Hourly_wages/hourly_wages.csv*" como dataframe utilizando a biblioteca Pandas e mostrar as 5 primeiras e 5 últimas linhas.

```
!git clone https://github.com/malegopc/DSBD

import pandas as pd

df = pd.read_csv('/content/DSBD/Datasets/Hourly_wages/hourly_wages.csv')

print("Primeiras 5 linhas:")
print(df.head())

print("\nÚltimas 5 linhas:")
print(df.tail())
```

#### c-) Mostrar os histogramas dos atributos: *education_yrs*, *experience_yrs* e *age* num mesmo quadro de plotagem (usando *subplots*).

```
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

df['education_yrs'].plot(kind='hist', ax=axes[0], title='Education Years')
df['experience_yrs'].plot(kind='hist', ax=axes[1], title='Experience Years')
df['age'].plot(kind='hist', ax=axes[2], title='Age')

plt.tight_layout()
plt.show()
```

#### d-) Mostrar os gráficos de setores (*pie chart*) de cada um dos atributos: *union*, *female*, *marr*, *south*, *manufacturing* e *construction* num mesmo quadro de plotagem (usando subplots).

```
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

df['union'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[0, 0], title='Union')
df['female'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[0, 1], title='Female')
df['marr'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[0, 2], title='Married')
df['south'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[1, 0], title='South')
df['manufacturing'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[1, 1], title='Manufacturing')
df['construction'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[1, 2], title='Construction')

plt.tight_layout()
plt.show()
```

#### e-) Mostrar os boxplots das variáveis *education_yrs*, *experience_yrs* e *age* num mesmo quadro de plotagem (usando *subplots*).

```
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

df['education_yrs'].plot(kind='box', ax=axes[0], title='Education Years')
df['experience_yrs'].plot(kind='box', ax=axes[1], title='Experience Years')
df['age'].plot(kind='box', ax=axes[2], title='Age')

plt.tight_layout()
plt.show()
```

#### f-) Mostre lado a lado os boxplots da variável *age* discriminados por sexo (female).

```
import matplotlib.pyplot as plt

df.boxplot(column='age', by='female', figsize=(8, 6))
plt.title('Boxplot de Idade por Sexo')
plt.suptitle('')
plt.xlabel('Sexo (0 = Male, 1 = Female)')
plt.ylabel('Idade')
plt.show()
```

#### h-) Mostre lado a lado os boxplots da variável *wage_per_hour* discriminados por sexo (female).

```
import matplotlib.pyplot as plt
df.boxplot(column='wage_per_hour', by='female', figsize=(8, 6))
plt.title('Boxplot de Salário por Hora por Sexo')
plt.suptitle('')
plt.xlabel('Sexo (0 = Male, 1 = Female)')
plt.ylabel('Salário por Hora')
plt.show()
```

Resposta:os homens possuem uma mediana maior,a dispersao e maior para os homens, as mulheres possuem mais outliners que os homens mesmo tendo uma media menor.

#### j-) Mostre lado a lado os boxplots da variável wage_per_hour discriminados por estado civil (marr).

```
import matplotlib.pyplot as plt
df.boxplot(column='wage_per_hour', by='marr', figsize=(8, 6))
plt.title('Boxplot de Salário por Hora por Estado Civil')
plt.suptitle('')
plt.xlabel('Estado Civil (0 = Solteiro, 1 = Casado)')
plt.ylabel('Salário por Hora')
plt.show()
```
